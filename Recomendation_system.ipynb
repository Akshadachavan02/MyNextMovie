```python
import pandas as pd
```

```python
rating_df = pd.read_csv('u.data' ,sep='\t',names=['user id','term id','rating','timestamp'])
movies_df = pd.read_csv('u.item',
                        sep="|",
                       names=['movie_id', 'movie_title', 'release_date', 
                              'video_release_date', 'IMDb URL', 'unknown',
                              'Action', 'Adventure', 'Animation', 'Children', 
                            'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
                            'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 
                                     'Thriller', 'War', 'Western'],encoding = "latin-1")
movies_df = movies_df.iloc[:,[0,1]]
```

```python
rating_df.head()
```

```python
movies_df.head()
```

```python
# lets build a association matrix 
learningmat = rating_df.pivot_table(index=["term id"],
                                    columns =["user id"],values = "rating").reset_index(drop=True)
learningmat.head()
```

```python
learningmat.shape
```

```python
learningmat.fillna(0,inplace=True)
learningmat.head()
```

```python
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
movie_similarity = pairwise_distances(learningmat, metric = "cosine")
np.fill_diagonal(movie_similarity,0)
ratings_matrix = pd.DataFrame(movie_similarity)
ratings_matrix.head()
```

```python
ratings_matrix.shape
```

```python
try:
    user_no = input("Enter the reference movie title based on which recommendations are to be made:")
    inp = movies_df[movies_df['movie_title']==user_no].index.tolist()
    inp = inp[0]
    movies_df["similarity"] = ratings_matrix.iloc[inp]
    print("Recommended movies based on your choice of ",user_inp ,": \n", movies_df.sort_values( ["similarity"],
                                                                                                 ascending = False )[1:10])
    
except:
    print("The movie name you have entered does not exist in the list, however, below are the top movies recommended in general")
    movies_df= movies_df.drop("similarity", axis=1)
    print(movies_df.head(10))
```
